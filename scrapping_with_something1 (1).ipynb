```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
urll = "http://quotes.toscrape.com/"
re = requests.get(urll)
re
```

```python
data = BeautifulSoup(re.text,"html.parser")
data
```

```python
title = data.find("a",style="text-decoration: none")
print(title.text.strip())
```

```python
all = []
for page in range(1, 3):
    url = f"http://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    qu = soup.find_all("span", class_="text")
    for i in qu:
        all.append(i.text.strip())
print(all)
```

```python
author = []
for page_num in range(1, 3):
    url = f"http://quotes.toscrape.com/page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    author1 = soup.find_all("small", class_="author")
    for i in author1:
        author.append(i.text.strip())
author
```

```python
tags = []
for page_num in range(1, 3):
    url = f"http://quotes.toscrape.com/page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    tags1 = soup.find_all("div", class_="tags")
    for j in tags1:
        tags.append(j.text.replace("\n", ",").replace(",            Tags:,            ,", "").strip())
tags
```

```python
df1 = pd.DataFrame({"Ideas":all,"authors":author,"tags":tags})
df1
```

```python
df1.to_csv("quotes.csv")
```
